```python
# import dependencies

import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Merging of the detailed CSVs which contain individual posts, their corresponding positive, negative and neutral sentiment scores, and the candidate the sentiment refers to.

##### 2016 Dataset cleaning and preparation for merge

```python
# Read in 2016 CSV file
ind_post_data_2016_df = pd.read_csv("CSV_Output_Sentiment/sent_and_cand_by_post_2016.csv",)

# Display sample data
ind_post_data_2016_df.head(5)
```

```python
# Drop columns that start with 'token_'
columns_to_drop = [col for col in ind_post_data_2016_df.columns if col.startswith('token_')]
ind_post_data_2016_df = ind_post_data_2016_df.drop(columns=columns_to_drop)

# Verify the result
ind_post_data_2016_df.head()
```

```python
# Add a column 'Year' with a date value for each row
ind_post_data_2016_df['Year'] = 2016

# Verify the result
ind_post_data_2016_df.head()
```

```python
# Rename the columns
ind_post_data_2016_df = ind_post_data_2016_df.rename(columns={
    'text': 'Text',
    'candidate': 'Candidate'
})

# Display the updated DataFrame
ind_post_data_2016_df.head()
```

##### 2020 Dataset cleaning and preparation for merge

```python
# Read in 2020 CSV file
ind_post_data_2020_df = pd.read_csv("CSV_Output_Sentiment/sent_and_cand_by_post_2020.csv",)

# Display sample data
ind_post_data_2020_df.head(5)
```

```python
# Define the desired order of columns
desired_columns = ['Compound', 'Positive', 'Negative', 'Neutral', 'text', 'candidate']

# Select and reorder the DataFrame columns
ind_post_data_2020_df = ind_post_data_2020_df[desired_columns]

# Display the resulting DataFrame
ind_post_data_2020_df.head()
```

```python
# Add a column 'Year' with a date value for each row
ind_post_data_2020_df['Year'] = 2020

# Verify the result
ind_post_data_2020_df.head()
```

```python
# Rename the columns
ind_post_data_2020_df = ind_post_data_2020_df.rename(columns={
    'text': 'Text',
    'candidate': 'Candidate'
})

# Display the updated DataFrame
ind_post_data_2020_df.head()
```

```python
# Replace values in the 'Candidate' column
ind_post_data_2020_df['Candidate'] = ind_post_data_2020_df['Candidate'].replace({
    'democrat': 'Democrat',
    'republican': 'Republican'
})

# Display the updated DataFrame
ind_post_data_2020_df.head()
```

##### 2024 Dataset cleaning and preparation for merge

```python
# Read in 2024 CSV file
ind_post_data_2024_df = pd.read_csv("CSV_Output_Sentiment/sent_and_cand_by_post_2024.csv",)

# Display sample data
ind_post_data_2024_df.head(5)
```

```python
# Drop columns that start with 'token_'
columns_to_drop = [col for col in ind_post_data_2024_df.columns if col.startswith('token_')]
ind_post_data_2024_df = ind_post_data_2024_df.drop(columns=columns_to_drop)

# Filter out rows where the 'Candidate' column is not 'Republican' or 'Democrat'
valid_candidates = ['republican', 'democrat']
ind_post_data_2024_df = ind_post_data_2024_df[ind_post_data_2024_df['candidate'].isin(valid_candidates)]

# Verify the result
ind_post_data_2024_df.head()
```

```python
# Add a column 'Year' with a date value for each row
ind_post_data_2024_df['Year'] = 2024

# Verify the result
ind_post_data_2024_df.head()
```

```python
# Rename the columns
ind_post_data_2024_df = ind_post_data_2024_df.rename(columns={
    'text': 'Text',
    'candidate': 'Candidate'
})

# Display the updated DataFrame
ind_post_data_2024_df.head()
```

```python
# Replace values in the 'Candidate' column
ind_post_data_2024_df['Candidate'] = ind_post_data_2024_df['Candidate'].replace({
    'democrat': 'Democrat',
    'republican': 'Republican'
})

# Display the updated DataFrame
ind_post_data_2024_df.head()
```

##### Concatenate the 3 dataframes into a single dataframe

```python
# Concatenate them into a single DataFrame
final_ind_post_data_df = pd.concat([ind_post_data_2016_df, ind_post_data_2020_df, ind_post_data_2024_df], ignore_index=True)

# Verify the result
print(f"Total rows in combined DataFrame: {len(final_ind_post_data_df)}")
final_ind_post_data_df.head()
```

```python
# Save the final combined individual post data dataFrame to a CSV file
final_ind_post_data_df.to_csv('../CSV_Outputs/final_ind_post_data_df.csv', index=False)
```

### 2. Merging of the summary CSVs which contain the average sentiment scores per candidate for the three years of available data.

##### 2016 Dataset cleaning and preparation for merge

```python
# Read in 2016 CSV file
summary_data_2016_df = pd.read_csv("CSV_Output_Sentiment/sent_info_for_db_2016.csv",)

# Display sample data
summary_data_2016_df.head(5)
```

```python
# Rename the columns
summary_data_2016_df = summary_data_2016_df.rename(columns={
    'id': 'ID',
    'text': 'Text',
    'time': 'Year',
    'candidate': 'Candidate',
    'average_sentiment': 'Avg_Sentiment'
})

# Display the updated DataFrame
summary_data_2016_df.head()
```

```python
# Replace values in the 'Candidate' column
summary_data_2016_df['Candidate'] = summary_data_2016_df['Candidate'].replace({
    'democrat': 'Democrat',
    'republican': 'Republican'
})

# Display the updated DataFrame
summary_data_2016_df.head()
```

##### 2020 Dataset cleaning and preparation for merge

```python
# Read in 2020 CSV file
summary_data_2020_df = pd.read_csv("CSV_Output_Sentiment/sent_info_for_db_2020.csv",)

# Display sample data
summary_data_2020_df.head(5)
```

```python
# Rename the columns
summary_data_2020_df = summary_data_2020_df.rename(columns={
    'id': 'ID',
    'text': 'Text',
    'time': 'Year',
    'candidate': 'Candidate',
    'average_sentiment': 'Avg_Sentiment'
})

# Display the updated DataFrame
summary_data_2020_df.head()
```

```python
# Replace values in the 'Candidate' column
summary_data_2020_df['Candidate'] = summary_data_2020_df['Candidate'].replace({
    'democrat': 'Democrat',
    'republican': 'Republican'
})

# Display the updated DataFrame
summary_data_2020_df.head()
```

##### 2024 Dataset cleaning and preparation for merge

```python
# Read in 2024 CSV file
summary_data_2024_df = pd.read_csv("CSV_Output_Sentiment/sent_info_for_db_2024.csv",)

# Display sample data
summary_data_2024_df.head(5)
```

```python
# Rename the columns
summary_data_2024_df = summary_data_2024_df.rename(columns={
    'id': 'ID',
    'text': 'Text',
    'time': 'Year',
    'candidate': 'Candidate',
    'average_sentiment': 'Avg_Sentiment'
})

# Display the updated DataFrame
summary_data_2024_df.head()
```

```python
# Replace values in the 'Candidate' column
summary_data_2024_df['Candidate'] = summary_data_2024_df['Candidate'].replace({
    'democrat': 'Democrat',
    'republican': 'Republican'
})

# Display the updated DataFrame
summary_data_2024_df.head()
```

##### Concatenate the 3 dataframes into a single dataframe

```python
# Concatenate them into a single DataFrame
final_sent_summary_data_df = pd.concat([summary_data_2016_df, summary_data_2020_df, summary_data_2024_df], ignore_index=True)

# Remove rows where the 'Candidate' column is 'unknown'
final_sent_summary_data_df = final_sent_summary_data_df[final_sent_summary_data_df['Candidate'].str.lower() != 'unknown']

# Verify the result
print(f"Total rows in combined DataFrame: {len(final_sent_summary_data_df)}")
final_sent_summary_data_df.head()
```

```python
# Save the final combined individual post data dataFrame to a CSV file
final_sent_summary_data_df.to_csv('../CSV_Outputs/final_sentiment_summary_data_df.csv', index=False)
```

### Visualizations

```python
# Set up the plot
plt.figure(figsize=(12, 8))

# Define a custom color palette
color_palette = {
    'Democrat': 'blue',
    'Republican': 'red'
}

# Create the diverging bar plot with a custom color palette
sns.barplot(
    data=final_sent_summary_data_df,
    x='Year',
    y='Avg_Sentiment',
    hue='Candidate',
    palette=color_palette,  # Use the custom color palette
    dodge=True
)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Average Sentiment Score')
plt.title('Average Sentiment Score by Candidate and Year')

# Add a horizontal line at y=0 for better visual comparison
plt.axhline(0, color='gray', linestyle='--')

# Set y-axis limits
plt.ylim(-0.10, 0.50)

# Adjust x-ticks if needed
plt.xticks

# Show the plot
plt.tight_layout()
plt.savefig('../Visualizations/all_years_avg_sent_by_cand.png')
plt.show()
```
